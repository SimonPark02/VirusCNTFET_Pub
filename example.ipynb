# Numerical simulation with TDS model

```python
from tds_model import VoltageResponse
from curves import create_step_voltage, create_voltage_sweep
```

Example 1. Response to step voltage

```python
VoltageResponse(
    *create_step_voltage(
        v_step=-80,
        v_sd=10.0,
        t_polarization=100.0,
        polarization_delay=1.0
    ),
    t_max=500.0
).plot()
```

Example 2. Response to voltage sweep

```python
VoltageResponse(
    *create_voltage_sweep(
        v_low=-200.0,
        v_high=100.0,
        slope=300.0 / 40000,
        v_sd=10.0,
        t_polarization=100.0,
        polarization_delay=1.0
    )
).plot()
```
